# Pre-Processing of the image

```python
img = image.load_img(filename, target_size = (224,224))
```

```python
plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
```

```python
resized_img   #Konsep dari image classification adalah mengubah suatu gambar menjadi barisan array yang memiliki value
```

```python
final_image = np.expand_dims(resized_img,axis = 0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
resized_img.shape #karena hanya 3 dimensi maka harus membutuhkan 4 dimensi
```

```python
final_image.shape
```

```python
predictions = mobile.predict(final_image)
```

```python
predictions
```

```python
from tensorflow.keras.applications import imagenet_utils  #Untuk DECODING
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```

```python
plt.imshow(img)
```
